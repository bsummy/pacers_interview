```python
import pandas as pd
import matplotlib.pyplot as plt

pacers_data = pd.read_csv('pacers_results_four_factors.csv')
league_average_data = pd.read_csv('pacers_results_four_factors.csv')

factors = [('efg_pct', 'opp_efg_pct'), ('oreb_pct', 'opp_oreb_pct'), ('tov_pct','opp_tov_pct'), ('ftm_rate', 'opp_ftm_rate')]
tiny_image_path = '/Users/bennettsummy/pacers_interview/q2/pacers.png'

wins_data = pacers_data[pacers_data['win'] == True]
losses_data = pacers_data[pacers_data['win'] == False]

for factor in factors:
    plt.figure(figsize=(10, 6))

    plt.scatter(wins_data[factor[0]], wins_data[factor[1]], label='Wins', marker='o', color = 'g')
    plt.scatter(losses_data[factor[0]], losses_data[factor[1]], label='Losses', marker='x', color = 'r')
    plt.axhline(y=league_average_data[factor[0]].iloc[0], color='r', linestyle='--', label='League Avg')

    plt.title(f'2022-2023 Pacers {factor[0]} vs Opponent {factor[0]}')
    plt.xlabel('Pacers ' + factor[0])
    plt.ylabel('Opponent ' + factor[0])
    plt.legend()
    plt.show()
```

# All charts seem inconclusive.
Over-analysis of data is just as dangerous as under-analysis. However, the most descriptive statistic seems to be eFG - when the Pacers shoot well, and the opponents shoot poorly, the Pacers tend to win.
